```python
import gzip, json, os, time, pickle
import pandas as pd

# directory where data is stored
load_DIR = '/media/mthommes/MThommes/Insight/'
# directory where data will be saved
save_DIR = '/home/mthommes/insight/project/'
```

# Functions

```python
def save_data(load_name, save_name):
    """
    save_data counts the number of lines in a json.gz file
    INPUTS
    load_name: full path of file to read (str)
    save_name: full path of file to save (str)
    """
    # figure out how many data points to read in
    n = count_data(load_name)
    print('There are', n, 'data points')

    # read in the data
    data, count = read_data(os.path.join(load_DIR, load_name), n)
    print('Read', count, 'data points')

    # save to file
    data.to_pickle(save_name)
    print('Saved data')    

def count_data(file_name):
    """
    count_data counts the number of lines in a json.gz file
    INPUT
    file_name: full path of file to read (str)
    OUTPUT
    count: number of lines (int)
    """
    with gzip.open(file_name) as f_open:
        count = 0
        for line in f_open:
            count += 1
    return count

def read_data(file_name, n=10):
    """
    read_data reads the data in a json.gz file
    INPUTS
    file_name: full path of file to read (str)
    n: number of lines to read (int)
    OUTPUT
    df: read data (DataFrame)
    count: number of lines read (int)
    """
    with gzip.open(file_name) as f_open:
        count = 0
        col_names = get_col_names(file_name)
        df = pd.DataFrame(columns=col_names)
        for line in f_open:
            d = json.loads(line)
            d_sr = pd.Series
            d_sr = pd.Series(d, index=col_names)
            df = df.append(d_sr, ignore_index=True)
            # increase iteration number
            count += 1
            # break if reaches the 10th (default) line
            if (n is not None) and (count >= n):
                break
            if count % 1E3 == 0:
                print(count)
    return df, count

def get_col_names(file_name):
    """
    get_col_names gets the keys in a json.gz file
    INPUT
    file_name: full path of file to read (str)
    OUTPUT
    col_names: keys (list)
    """
    with gzip.open(file_name) as f_open:
        for line in f_open:
            d = json.loads(line)
            col_names = list(d.keys())
            break
    return col_names
```
